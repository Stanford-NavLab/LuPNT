```python
import pylupnt as pnt
import numpy as np
```

```python
a = 6541.4
e = 0.6
i = np.deg2rad(65.5)
Omega = np.deg2rad(90.0)
w = np.deg2rad(0.0)
M = np.deg2rad(0.0)

mu = pnt.MU_MOON
coe_state = pnt.ClassicalOE(a, e, i, Omega, w, M, cs=pnt.CoordSystem.MI)
rv_state = pnt.coe_to_cart(coe_state, mu)
```
