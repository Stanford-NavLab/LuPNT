```python
import pylupnt as pnt
import numpy as np
```

## Set the initial orbit elements of the orbit

```python
p = 11067.790
e = 0.83285
a = p / (1 - pow(e, 2))     
inc = np.deg2rad(87.87)
Omega = np.deg2rad(227.89)
w = np.deg2rad(53.38)
nu = np.deg2rad(92.335)
M = pnt.true_to_mean_anomaly(nu, e)

oe_array = np.array([a, e, inc, Omega, w, M])
```

### Setup the Orbit Elements 
The first step is to setup the states of the Orbital Elements
For Keplarian orbit elements (classical, equinoctial, QuasiNonsingular), you can propagate the orbit by directly updating the state class in the propagator

```python
# Orbit Elements (Classical)
class_oe = pnt.ClassicalOE(oe_array, frame=pnt.Frame.MI)  # classical orbital elements class (the second argument is the frame of reference)
x_oe = class_oe.vector
print(" ")
print("Classical orbital elements:")
print(x_oe)

# Orbital Elements (Quasi-Nonsingular)
class_qn = pnt.QuasiNonsingularOE(oe_array, frame=pnt.Frame.MI)  # quasi-nonsingular orbital elements class
x_qn = class_qn.vector
print(" ")
print("Quasi-nonsingular orbital elements:")
print(x_qn)
```

## Setup the Dynamics
We propagate the orbit using three different types of dynamics types

### Keplarian Dynamics
This dynamics class propagates the orbit elements in a two-setting. No perturbations are added

```python
# dynamics
mu_moon = pnt.MU_MOON
keplerian_dynamics = pnt.KeplerianDynamics(mu_moon)
```

We will set the propagation time epoch to a single orbital period

```python
T_orbit = 2 * np.pi * np.sqrt(pow(a, 3) / mu_moon)
print(" Orbital period: ", T_orbit, " seconds")

# Propagate the orbit for half an orbit
t_end = 1 * T_orbit

# 1st method: Directly updates the orbit elements class
print("State at t0:")
print(class_oe.vector)
keplerian_dynamics.propagate(class_oe, t_end)
print(" ")
print("State at t_end:")
print(class_oe.vector)
```
